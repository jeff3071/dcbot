```python
import csv
import json
```

```python
with open('gf.txt', newline = '\n') as data:
    a = data.readlines()
```

```python
data_list = []
for i in a:
    li = i.split("/")
    data_list.append(li[len(li)-1].replace("\r\n",""))
```

```python
for x in range(len(data_list)):
    data_list[x]= data_list[x].split(".")[0]
```

```python
for x in range(len(data_list)):
    data_list[x]= data_list[x].split("_")[0]
```

```python
data_list
```

```python
with open('gf.json', 'w') as gf:
    json.dump(data_list, gf)
```
